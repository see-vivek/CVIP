# **Project Summary -**

The Global Terrorism Database (GTD) is an open-source database including information on terrorist attacks around the world from 1970 through 2017. The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

```python
# plotting bar graph for types of weapon used.
plt.figure(figsize=(15,5))
sns.countplot(x = df.Weapon_type)
plt.xticks(rotation=90)
plt.title('Weapons used for attack')
plt.show()
```

##### 1. What is/are the insight(s) found from the chart?

- Maximum number of attacks were done with Explosives followed by Firearms.
- least number of attacks were done  with Biological weapons.

#### Chart - 2

```python
#-Analysing year wise attack.
plt.figure(figsize=(15,5))
sns.countplot(x = df.	Year)
plt.xticks(rotation=90)
plt.title('year wise attack')
plt.show()
```

##### 1. What is/are the insight(s) found from the chart?

- Maximum number of attacks were done in the year 2014 followed by 2015.
- Minimum number of attacks were done in the year 1971 followed by 1973.

#### Chart - 3

```python
# analysing month wise attack.
plt.figure(figsize=(15,5))
sns.countplot(x = df.Month)
plt.xticks(rotation=90)
plt.title('Month wise attack')
plt.show()
```

##### 1. What is/are the insight(s) found from the chart?

- Maximum number of attack was done in the month of may.
- Minimum number of attack was done in the month of february and december.
- Almost every month attacks are equal.

#### Chart - 4

```python
# analysing day wise attack.
plt.figure(figsize=(15,5))
sns.countplot(x = df.Day)
plt.xticks(rotation=90)
plt.title('Day wise attack')
plt.show()
```

#### Chart - 5

```python
# analysing the attack country wise
plt.figure(figsize=(20,5))
sns.countplot(x = df.Country_name)
plt.xticks(rotation=90)
plt.rcParams.update({'font.size':4})
plt.title('Country_name wise attack')
plt.show()
```

##### 2. What is/are the insight(s) found from the chart?

#### Chart - 6
